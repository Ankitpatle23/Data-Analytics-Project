```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style='whitegrid')
```

```python
# Load the CSV file
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Check for any missing (null) values
df.isnull().sum()
```

```python
df.describe()
```

```python
# Count how many said 'Yes' and 'No' to Attrition
attrition_counts = df['Attrition'].value_counts()
```

```python
# Calculate percentage of employees who left the company
attrition_rate = attrition_counts['Yes'] / attrition_counts.sum() * 100
print(f"Overall Attrition Rate: {attrition_rate:.2f}%")
```

```python
# Plot number of attritions in each department
plt.figure(figsize=(8,5))  # Set size of plot (8 inches wide, 5 inches high)
sns.countplot(data=df[df['Attrition'] == 'Yes'], x='Department', palette='Set2')
plt.title("Attrition Count by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45)    # Rotate labels for readability
plt.show()
```

```python
# Count number of males and females
gender_counts = df['Gender'].value_counts()
```

```python
# Plot a pie chart of gender distribution
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue','lightpink'])
plt.title("Gender Distribution")
plt.show()
```

```python
# Plot average job satisfaction in each department
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Department', y='JobSatisfaction', palette='pastel', ci=None)
plt.title("Average Job Satisfaction by Department")
plt.ylabel("Job Satisfaction (1-4)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot how long employees worked vs attrition status
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', multiple='stack', palette='Set1', bins=15)
plt.title("Years at Company by Attrition")
plt.xlabel("Years at Company")
plt.ylabel("Employee Count")
plt.show()
```

```python
# Create a table showing counts between performance and work-life balance
heatmap_data = pd.crosstab(df['PerformanceRating'], df['WorkLifeBalance'])
```

```python
# Visualize the matrix as a heatmap
plt.figure(figsize=(6,4))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu')
plt.title("Performance Rating vs Work-Life Balance")
plt.xlabel("Work-Life Balance")
plt.ylabel("Performance Rating")
plt.show()
```

```python
# Print key findings from the analysis
insights = [
    "1. Overall attrition rate is {:.2f}%.".format(attrition_rate),
    "2. Highest attrition seen in the Sales department.",
    "3. Gender distribution is fairly balanced.",
    "4. Job satisfaction is lowest in [your observed department].",
    "5. Most employees leave within the first 3 years.",
    "6. Better work-life balance leads to better performance ratings."
]

print("\n".join(insights))
```
